# Data Preprocessing

```python
"""
Ufoh is a chatbot that applies a numeric grading to an input target message that optionally has up to five context messages (messages directly prior to the target) and a message being specifically replied to.

True/false variables for hate against race, religion, origin, gender, sexuality, age, and disabilities are also provided in the output.

Supportive range: -5 < 0
Neutral: 0
Hateful range: 0 < 5

The target message is specified by starting with "Message To Evaluate", while the messages previously said by OTHER PEOPLE is under CONTEXT ---, where "Message Being Replied To" is a directly reply. However, note that just because there is not a specific message being replied to doesn't mean that the previous messages are not indirectly being replied to. Therefore, take the context of the previous messages into account as well.

Format the output in this way:
hate_speech_score: #
target_race: true/false
target_religion: true/false
target_origin: true/false
target_gender: true/false
target_sexuality: true/false
target_age: true/false
target_disability: true/false

Example input:
CONTEXT ---
Message n-5: Hey guys
Message n-4: Welcome
Message n-3: Hii
Message n-2: Welcome in bro
Message n-1: Yessir
Message Being Replied To: N/A
TARGET ---
Message To Evaluate: Shut up!

Example output:
hate_speech_score: 1
target_race: False
target_religion: False
target_origin: False
target_gender: False
target_sexuality: False
target_age: False
target_disability: False

Note that hate_speech_score is POSITIVE because the more POSITIVE the score, the MORE hateful the message.
"""

from better_profanity import profanity
profanity.load_censor_words()

# DATA PREPROCESSING // Format to jsonl, filter profanity (because direct slurs/profanity would not make it past a preset filter, OpenAI policy on hate speech)

role = "Ufoh is a chatbot that applies a numeric grading to an input target message that optionally has up to five context messages (messages directly prior to the target) and a message being specifically replied to. The more hateful a target message is, the more positive the number, while the more supportive the message is, the more negative the number. True/false variables for hate against race, religion, origin, gender, sexuality, age, and disabilities are also provided in the output. The target message is specified by starting with TARGET ---, while the messages previously said by other people is under CONTEXT ---, where Message Being Replied To is being directly responded to. However, note that just because there is not a specific message being replied to doesn't mean that the previous messages are not indirectly being replied to. Therefore, take the context of the previous messages into account as well. Format the output in this way: hate_speech_score: # target_race: true/false target_religion: true/false target_origin: true/false target_gender: true/false target_sexuality: true/false target_age: true/false target_disability: true/false"

for i in range(1000, 1001):
    target = df.loc[i, 'text']
    target = target.replace('"', '\\"')
    target = profanity.censor(target)
    input = "CONTEXT --- Message n-5: N/A Message n-4: N/A Message n-3: N/A Message n-2: N/A Message n-1: N/A Message Being Replied To: N/A TARGET --- Message To Evaluate: " + target
    output = "hate_speech_score: " + str(df.loc[i, 'hate_speech_score']) + " target_race: " + str(df.loc[i, 'target_race']) + " target_religion: " + str(df.loc[i, 'target_religion']) + " target_origin: " + str(df.loc[i, 'target_origin']) + " target_gender: " + str(df.loc[i, 'target_gender']) + " target_sexuality: " + str(df.loc[i, 'target_sexuality']) + " target_age: " + str(df.loc[i, 'target_age']) + " target_disability: " + str(df.loc[i, 'target_disability'])

    jsonl = "{\"messages\": [{\"role\": \"system\", \"content\": \"" + role + ".\"}, {\"role\": \"user\", \"content\": \"" + input + "\"}, {\"role\": \"assistant\", \"content\": \"" + output + "\"}]}"
    print(jsonl) # Used in the .jsonl file to fine-tune GPT 3.5

    # Used for evaluating performance
    #print(input)
    #print(output)
```

# Performance Measures

```python
import openai
import os
from openai import OpenAI
from google.colab import userdata
import numpy as np
from sklearn import metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Function to determine if the output matches the desired format
def is_correct_format(output):
    required_fields = [
        "hate_speech_score", "target_race", "target_religion",
        "target_origin", "target_gender", "target_sexuality",
        "target_age", "target_disability"
    ]
    for field in required_fields:
        if f"{field}:" not in output:
            return False
    return True

os.environ["OPENAI_API_KEY"] = userdata.get('openapikey')
client = OpenAI()

role = ("Ufoh is a chatbot that applies a numeric grading to an input target message that optionally has up to five "
        "context messages (messages directly prior to the target) and a message being specifically replied to. "
        "True/false variables for hate against race, religion, origin, gender, sexuality, age, and disabilities are "
        "also provided in the output. Supportive range: -5 < 0, Neutral: 0, Hateful range: 0 < 5. The target message is "
        "specified by starting with 'Message To Evaluate', while the messages previously said by OTHER PEOPLE is under "
        "CONTEXT ---, where 'Message Being Replied To' is a directly reply. However, note that just because there is "
        "not a specific message being replied to doesn't mean that the previous messages are not indirectly being "
        "replied to. Therefore, take the context of the previous messages into account as well. Format the output in "
        "this way: hate_speech_score: #, target_race: true/false, target_religion: true/false, target_origin: true/false, "
        "target_gender: true/false, target_sexuality: true/false, target_age: true/false, target_disability: true/false. "
        "Example input: CONTEXT --- Message n-5: Hey guys Message n-4: Welcome Message n-3: Hii Message n-2: Welcome in "
        "bro Message n-1: Yessir Message Being Replied To: N/A TARGET --- Message To Evaluate: Shut up! Example output: "
        "hate_speech_score: 1 target_race: False target_religion: False target_origin: False target_gender: False "
        "target_sexuality: False target_age: False target_disability: False. Note that hate_speech_score is POSITIVE "
        "because the more POSITIVE the score, the MORE hateful the message.")

# PERFORMANCE EVALUATION (CREATING CONFUSION MATRIX FOR EACH CATEGORY)
start = 1000
evalSize = 500
cont = 0
hs_truth = []
hs_pred = []
race_truth = []
race_pred = []
religion_truth = []
religion_pred = []
origin_truth = []
origin_pred = []
gender_truth = []
gender_pred = []
sexuality_truth = []
sexuality_pred = []
age_truth = []
age_pred = []
disability_truth = []
disability_pred = []

for i in range(start, start + evalSize):
  target = df.loc[i, 'text']
  input = "CONTEXT --- Message n-5: N/A Message n-4: N/A Message n-3: N/A Message n-2: N/A Message n-1: N/A Message Being Replied To: N/A TARGET --- Message To Evaluate: " + target

  completion = client.chat.completions.create(
  model="ft:gpt-3.5-turbo-0125:personal:ufoh-v3:9ETnxurq", # Call the fine-tuned pretrained model
  #model="gpt-4-turbo",
  messages=[
    {"role": "system", "content": role},
    {"role": "user", "content": input}
    ]
  )
  ai_output = str(completion.choices[0].message)
  #print(ai_output)

  # Try to find a valid output format (5 times max)
  valid_output_found = is_correct_format(ai_output)
  index = 0
  while not valid_output_found and index < min(5, len(completion.choices)):
      ai_output = str(completion.choices[index].message)
      valid_output_found = is_correct_format(ai_output)
      index += 1

  # Check if valid output was found and use it
  if not valid_output_found:
    continue

  # Extract the content part of the string
  try:
    # Attempt to parse the content string from ai_output
    content_str = ai_output.split("content='")[1].split("', role=")[0]
  except IndexError:
    continue

  # Initialize values
  ai_hate_speech_score = None
  ai_target_race = None
  ai_target_religion = None
  ai_target_origin = None
  ai_target_gender = None
  ai_target_sexuality = None
  ai_target_age = None
  ai_target_disability = None

  # Assign each parsed keyword to its value
  key_values = content_str.split()
  current_key = None
  for part in key_values:
      if ':' in part:
          if current_key is not None:
              value = ' '.join(temp_value).strip()
              if current_key == 'hate_speech_score':
                  decimal_found = False
                  res = ""
                  for ele in value:
                      if ele.isdigit():
                          res += ele
                      elif ele == '-' and not res:
                          res += ele
                      elif ele == '.' and not decimal_found:  # Only want one decimal (can be a period sometimes after the initial decimal)
                          res += ele
                          decimal_found = True
                  try:
                    ai_hate_speech_score = float(res)
                  except ValueError:
                    cont = 1
                    break
              elif current_key == 'target_race':
                  ai_target_race = value == 'True'
              elif current_key == 'target_religion':
                  ai_target_religion = value == 'True'
              elif current_key == 'target_origin':
                  ai_target_origin = value == 'True'
              elif current_key == 'target_gender':
                  ai_target_gender = value == 'True'
              elif current_key == 'target_sexuality':
                  ai_target_sexuality = value == 'True'
              elif current_key == 'target_age':
                  ai_target_age = value == 'True'
              elif current_key == 'target_disability':
                  ai_target_disability = value == 'True'

          current_key, temp_value = part.split(':')[0], [part.split(':')[1]]
      else:
          temp_value.append(part)
  if cont == 1:
    cont = 0
    continue
  # Process the last key-value if not processed yet
  if current_key is not None:
      value = ' '.join(temp_value).strip()
      if current_key == 'hate_speech_score':
          ai_hate_speech_score = float(value)
      elif current_key == 'target_race':
          ai_target_race = value == 'True'
      elif current_key == 'target_religion':
          ai_target_religion = value == 'True'
      elif current_key == 'target_origin':
          ai_target_origin = value == 'True'
      elif current_key == 'target_gender':
          ai_target_gender = value == 'True'
      elif current_key == 'target_sexuality':
          ai_target_sexuality = value == 'True'
      elif current_key == 'target_age':
          ai_target_age = value == 'True'
      elif current_key == 'target_disability':
          ai_target_disability = value == 'True'

  #print(ai_hate_speech_score, ai_target_race, ai_target_religion, ai_target_origin, ai_target_gender, ai_target_sexuality, ai_target_age, ai_target_disability)

  #PERFORMANCE EVALUATION FOR HATE/NOT HATE
  if df.loc[i, 'hate_speech_score'] <= 0:
    hs_truth.append(0)
  else:
    hs_truth.append(1)
  if ai_hate_speech_score <= 0:
    hs_pred.append(0)
  else:
    hs_pred.append(1)

  #PERFORMANCE EVALUATION FOR TARGET_RACE
  if df.loc[i, 'target_race'] == True:
    race_truth.append(1)
  else:
    race_truth.append(0)
  if ai_target_race == True:
    race_pred.append(1)
  else:
    race_pred.append(0)

  #PERFORMANCE EVALUATION FOR TARGET_RELIGION
  if df.loc[i, 'target_religion'] == True:
    religion_truth.append(1)
  else:
    religion_truth.append(0)
  if ai_target_religion == True:
    religion_pred.append(1)
  else:
    religion_pred.append(0)

  #PERFORMANCE EVALUATION FOR TARGET_ORIGIN
  if df.loc[i, 'target_origin'] == True:
    origin_truth.append(1)
  else:
    origin_truth.append(0)
  if ai_target_origin == True:
    origin_pred.append(1)
  else:
    origin_pred.append(0)

  #PERFORMANCE EVALUATION FOR TARGET_GENDER
  if df.loc[i, 'target_gender'] == True:
    gender_truth.append(1)
  else:
    gender_truth.append(0)
  if ai_target_gender == True:
    gender_pred.append(1)
  else:
    gender_pred.append(0)

  #PERFORMANCE EVALUATION FOR TARGET_SEXUALITY
  if df.loc[i, 'target_sexuality'] == True:
    sexuality_truth.append(1)
  else:
    sexuality_truth.append(0)
  if ai_target_sexuality == True:
    sexuality_pred.append(1)
  else:
    sexuality_pred.append(0)

  #PERFORMANCE EVALUATION FOR TARGET_AGE
  if df.loc[i, 'target_age'] == True:
    age_truth.append(1)
  else:
    age_truth.append(0)
  if ai_target_age == True:
    age_pred.append(1)
  else:
    age_pred.append(0)

  #PERFORMANCE EVALUATION FOR TARGET_DISABILITY
  if df.loc[i, 'target_disability'] == True:
    disability_truth.append(1)
  else:
    disability_truth.append(0)
  if ai_target_disability == True:
    disability_pred.append(1)
  else:
    disability_pred.append(0)

  #print(hs_truth)
  #print(hs_pred)

# DISPLAY RESULTS FOR HATE/NOT HATE
confusion_matrix = metrics.confusion_matrix(hs_truth, hs_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["Not Hate Speech", "Hate Speech"])
hs_report = classification_report(hs_truth, hs_pred)
print(hs_report)
cm_display.plot()
plt.show()

# DISPLAY RESULTS FOR TARGET_RACE
confusion_matrix = metrics.confusion_matrix(race_truth, race_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["Does Not Target Race", "Targets Race"])
race_report = classification_report(race_truth, race_pred)
print(race_report)
cm_display.plot()
plt.show()

# DISPLAY RESULTS FOR TARGET_RELIGION
confusion_matrix = metrics.confusion_matrix(religion_truth, religion_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["Does Not Target Religion", "Targets Religion"])
religion_report = classification_report(religion_truth, religion_pred)
print(religion_report)
cm_display.plot()
plt.show()

# DISPLAY RESULTS FOR TARGET_ORIGIN
confusion_matrix = metrics.confusion_matrix(origin_truth, origin_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["Does Not Target Origin", "Targets Origin"])
origin_report = classification_report(origin_truth, origin_pred)
print(origin_report)
cm_display.plot()
plt.show()

# DISPLAY RESULTS FOR TARGET_GENDER
confusion_matrix = metrics.confusion_matrix(gender_truth, gender_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["Does Not Target Gender", "Targets Gender"])
gender_report = classification_report(gender_truth, gender_pred)
print(gender_report)
cm_display.plot()
plt.show()

# DISPLAY RESULTS FOR TARGET_SEXUALITY
confusion_matrix = metrics.confusion_matrix(sexuality_truth, sexuality_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["Does Not Target Sexuality", "Targets Sexuality"])
sexuality_report = classification_report(sexuality_truth, sexuality_pred)
print(sexuality_report)
cm_display.plot()
plt.show()

# DISPLAY RESULTS FOR TARGET_AGE
confusion_matrix = metrics.confusion_matrix(age_truth, age_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["Does Not Target Age", "Targets Age"])
age_report = classification_report(age_truth, age_pred)
print(age_report)
cm_display.plot()
plt.show()

# DISPLAY RESULTS FOR TARGET_DISABILITY
confusion_matrix = metrics.confusion_matrix(disability_truth, disability_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["Does Not Target Disability", "Targets Disability"])
disability_report = classification_report(disability_truth, disability_pred)
print(disability_report)
cm_display.plot()
plt.show()
```
